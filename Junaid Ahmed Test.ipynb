```python
class Book():
    
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_borrowed = False

    def borrow_book(self):
        if not self.is_borrowed_book:
            self.is_borrowed_book = True
            print(f"{self.title} has been borrowed from library.")
        else:
            print(f"{self.is_borrowed_book} is already borrowed from librabry.")

    def return_book(self):
        if self.is_borrowed_book:
            self.is_borrowed_book = False
            print(f"{self.title} has been returned to library.")
        else:
            print(f"{self.title} was not borrowed to library.")

class DigitalBook(Book):
    
    def __init__(self, title, author, isbn, file_format):
        super().__init__(title, author, isbn)
        self.file_format = file_format

    def borrow(self):
        super().borrow()
        if self.is_borrowed_book:
            print(f"You are allowed to access {self.title} online in {self.file_format} format.")

class AudioBook(Book):
    
    def __init__(self, title, author, isbn, duration):
        super().__init__(title, author, isbn)
        self.duration = duration

    def borrow(self):
        super().borrow()
        if self.is_borrowed_book:
            print(f"You are allowed to listen {self.title} for {self.duration} hours.")

class User:
    
    def __init__(self, user_id, name):
        self.user_id = user_id
        self.name = name
        self.is_borrowed_book = []

    def borrow_book(self, book):
        if not book.is_borrowed_book:
            book.borrow()
            self.is_borrowed_book.append(book)
        else:
            print(f"You are not allowed to borrow {book.title}: it is already borrowed from library.")

    def return_book(self, book):
        if book in self.is_borrowed_book:
            book.return_book()
            self.is_borrowed_book.remove(book)
        else:
            print(f"You are not allowed to return {book.title}: as it was not borrowed by you.")

class Library:
    
    def __init__(self, name):
        self.name = name
        self.book = []
        self.users = []

    def add_book(self, book):
        self.book.append(book)
        print(f"{book.title} has been added to the library.")

    def register_user(self, user):
        self.users.append(user)
        print(f"{user.name} has been registered as a user of library.")

    def lend_book(self, user_id, isbn):
        user = next((u for u in self.users if u.user_id == user_id), None)
        book = next((b for b in self.book if b.isbn == isbn), None)

        if user and book:
            user.borrow_book(book)
        else:
            print("Invalid User/Book or User and Book not found.")

    def receive_return(self, user_id, isbn):
        user = next((u for u in self.users if u.user_id == user_id), None)
        book = next((b for b in self.books if b.isbn == isbn), None)

        if user and book:
            user.return_book(book)
        else:
            print("User  or Book not found.")

# example
if __name__ == "__main__":
    library = Library("Town Library")

    book1 = Book("1996", "Jawaid Ahmed", "12345")
    book2 = DigitalBook("Fall of the Empire", "Muhammad Raheel", "24680", "PDF")
    book3 = AudioBook("Life", "Pervaiz Iqbal", "13579", 17.5)

    library.add_book(book1)
    library.add_book(book2)
    library.add_book(book3)

    user1 = User("001", "Zain")
    user2 = User("002", "Junaid")
```
